## Data Preprocessing

This notebook prepares the raw video game sales dataset for analysis and modeling. The preprocessing stage focuses on cleaning the data, correcting inconsistencies, handling missing values, and transforming variables into formats suitable for analysis. The output of this notebook is a cleaned and structured dataset that will be used for exploratory data analysis, statistical testing, and machine learning modeling in later stages of the project.

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns

games = pd.read_excel('C:/Users/ADMIN/Downloads/gamessales_project/dataset/VideoGames_Sales.xlsx')
games.head(10)
```

```python
games['console'] = games['console'].astype('category')
games['genre'] = games['genre'].astype('category')
```

## Data Cleansing
There seem to be a lot of missing cells, some cols have upward 50k+ NA but some only have a couple. We will need to handle each carefully with appropriate methods

```python
games = games[games['total_sales(mil)'].notna()]
sales_cols = [
    'na_sales(mil)',
    'jp_sales(mil)',
    'pal_sales(mil)',
    'other_sales(mil)'
]
# Fill missing sales values with 0, as missing sales likely indicate no sales in that region
games[sales_cols] = games[sales_cols].fillna(0)
games['developer'] = games['developer'].fillna('Unknown')
games['release_year'] = games['release_date'].dt.year
games['release_month'] = games['release_date'].dt.month
games.info()
```

```python
games['critic_score'] = games.groupby('genre')['critic_score'].transform(
    lambda x: x.fillna(x.median()))
games['release_year'] = games['release_year'].fillna(games['release_year'].median())
games['release_month'] = games['release_month'].fillna(0) #Month = 0 means unknown month
games.isna().sum()
```

```python
games['sales_sum'] = (
    games['na_sales(mil)'] +
    games['jp_sales(mil)'] +
    games['pal_sales(mil)'] +
    games['other_sales(mil)']
)

(games['total_sales(mil)'] - games['sales_sum']).describe()
```

## Preprocessing Notes

The dataset contains a large number of missing values, particularly in sales and critic score columns. Since the primary focus of this project is analyzing video game sales, rows without total_sales were removed from the dataset. Regional sales variables with missing values were filled with zero where appropriate. Critic scores were imputed using genre-based median values to preserve genre-level rating characteristics. Missing release dates were handled by imputing the release year with the dataset median and marking unknown months as zero. These preprocessing steps resulted in a cleaned dataset suitable for exploratory analysis, statistical testing, and machine learning modeling.
