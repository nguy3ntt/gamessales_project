# Statistical Testings
The previous Exploration Data Analytic have reveals several patterns in this dataset, now statistics will help answer are these patterns real, or could they happen by chance?

1. Test: Does critic score affect sales? (Pearson)
2. Test: Do genres have different sales distributions? (one-way ANOVA)
3. Test: Regional sales differences (Kruskal-Wallis for skewed data)
4. Test: Console effect on sales (one-way ANOVA)

```python
import pandas as pd
from scipy.stats import pearsonr
from scipy.stats import f_oneway
from scipy.stats import kruskal

# load dataset
games = pd.read_csv('C:/Users/ADMIN/Downloads/gamessales_project/dataset/cleaned/cleaned.csv')
games.head(10)
```

## Does critic score affect sales?
EDA showed that corr ≈ 0.18, lets formally test if the correlation is statistically significant:
Hypothesis
- H0: critic_score and total_sales are not correlated
- H1: critic_score and total_sales are correlated

```python
# Calculate Pearson correlation between critic_score and total_sales(mil)
corr, p = pearsonr(games['critic_score'], games['total_sales(mil)'])

print("Correlation:", corr)
print("p-value:", p)
```

The result shows a weak positive correlation (r ≈ 0.18) with an extremely small p-value, indicating that the relationship is statistically significant. However, the relatively low correlation coefficient suggests that critic scores alone explain only a small portion of the variation in sales performance, and other factors likely play a larger role in determining commercial success.

## Do genres have different sales distributions?
EDA showed big differences between genres, lets test whether those differences are statistically significant.
Hypothesis
- H0: all genres have the same average sales
- H1: at least one genre has a different average sales

```python
groups = [
    group['total_sales(mil)'].values
    for name, group in games.groupby('genre')
]

f_stat, p = f_oneway(*groups)

print("F-stat:", f_stat)
print("p-value:", p)
```

The test produced a large F-statistic and an extremely small p-value, indicating that there are statistically significant differences in sales between genres. This supports the earlier exploratory analysis showing that certain genres, such as Action, Sports, and Shooter games, generate substantially higher total sales than others.

## Regional sales differences
We previously observed that NA > PAL > JP in sales
Hypothesis
- H0: regional sales distributions are equal
- H1: at least one region differs

```python
stat, p = kruskal(
    games['na_sales(mil)'],
    games['jp_sales(mil)'],
    games['pal_sales(mil)']
)

print("Statistic:", stat)
print("p-value:", p)
```

We yielded a very large test statistic with a p-value effectively equal to zero, indicating strong evidence that sales distributions differ significantly between regions. This confirms that the global video game market exhibits clear regional variations in consumer demand.

## Console effect on sales
Test if platform affects sales.

- H0: consoles have equal mean sales
- H1: at least one console differs

```python
groups = [
    group['total_sales(mil)'].values
    for name, group in games.groupby('console')
]

f_stat, p = f_oneway(*groups)

print("F-stat:", f_stat)
print("p-value:", p)
```

The results show a high F-statistic and an extremely small p-value, suggesting that console platform has a significant impact on game sales. This aligns with earlier observations that certain console generations, such as the PlayStation 2, Xbox 360, and PlayStation 3, generated substantially higher total sales than others.

# Conclusion
The statistical tests confirm several patterns identified during the exploratory analysis. While critic scores show a statistically significant relationship with sales, the effect is relatively weak, indicating that game quality alone does not strongly determine commercial success. In contrast, genre, platform, and regional market differences all show significant effects on sales performance, suggesting that structural factors within the gaming industry play a much larger role in shaping sales outcomes.

Together, these findings highlight that market dynamics, platform popularity, and genre trends are key drivers of video game success, rather than critic reception alone.
